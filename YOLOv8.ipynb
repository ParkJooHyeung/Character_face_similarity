```python
import os
import glob
import cv2
from matplotlib import pyplot as plt
from collections import defaultdict


def load_label(label_path):
    with open(label_path, 'r') as f:
        contents = f.readlines()
    return [list(map(float, line.strip().split())) for line in contents]


def draw_boxes(image, boxes, color=(0, 255, 0), thickness=2):
    img = image.copy()
    for box in boxes:
        class_id, x_center, y_center, width, height = box
        x_min = int((x_center - width / 2) * img.shape[1])
        x_max = int((x_center + width / 2) * img.shape[1])
        y_min = int((y_center - height / 2) * img.shape[0])
        y_max = int((y_center + height / 2) * img.shape[0])

        cv2.rectangle(img, (x_min, y_min), (x_max, y_max), color, thickness)

    return img


image_folder = '/content/runs/detect/predict'
image_paths = glob.glob(image_folder + '/*.jpg')

class_count = defaultdict(int)
total_images = len(image_paths)

for image_path in image_paths:
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    label_path = os.path.join(os.path.dirname(image_path), 'labels', os.path.basename(image_path).replace('.jpg', '.txt'))

    if not os.path.exists(label_path):
        print(f"{image_path}Ïóê ÏùºÏπòÌïòÎäî Î†àÏù¥Î∏îÏù¥ ÏóÜÏäµÎãàÎã§. Í±¥ÎÑàÎúÅÎãàÎã§...")
        continue

    label_data = load_label(label_path)

    for box in label_data:
        class_id = int(box[0])
        class_count[class_id] += 1

    image_with_boxes = draw_boxes(image, label_data)

    plt.imshow(image_with_boxes)
    plt.show()

print(f"Ï¥ù Ïù¥ÎØ∏ÏßÄ Ïàò: {total_images}")
for class_id, count in class_count.items():
    percentage = (count / total_images) * 100
    print(f"ÌÅ¥ÎûòÏä§ {class_id}: {count}Í∞úÏùò Í∞ùÏ≤¥ Í∞êÏßÄÎê® ({percentage:.2f}%)")
```
